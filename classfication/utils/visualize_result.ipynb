```python
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt
import os
```

```python
data_dir = Path('../processed_data_with_smooth_masks_augmented/')
case_ids = [file for file in os.listdir(data_dir) if file.startswith('case_00000')]
for case_id in case_ids:
    case_output_dir = data_dir / case_id
    tumor_roi = np.load(case_output_dir / "tumor_roi.npy")
    kidney_tumor_roi = np.load(case_output_dir / "kidney_tumor_roi.npy")

    num_slices = tumor_roi.shape[0]
    num_figs = 8
    fig, axs = plt.subplots(1, num_figs, figsize=(16, 2))
    for j in range(4):
        if j < num_slices:
            axs[j].imshow(tumor_roi[j], cmap='gray')
            axs[j].set_title(f'Tumor ROI {j}')
        else:
            axs[j].imshow(np.zeros_like(tumor_roi[0]), cmap='gray')
            axs[j].set_title(f'Tumor ROI {j}')
        axs[j].axis('off')
    for j in range(4):
        if j < num_slices:
            axs[j+4].imshow(kidney_tumor_roi[j], cmap='gray')
            axs[j+4].set_title(f'Kidney+Tumor ROI {j}')
        else:
            axs[j+4].imshow(np.zeros_like(kidney_tumor_roi[0]), cmap='gray')
            axs[j+4].set_title(f'Kidney+Tumor ROI {j}')
        axs[j+4].axis('off')
    plt.tight_layout()
    plt.show()
```
